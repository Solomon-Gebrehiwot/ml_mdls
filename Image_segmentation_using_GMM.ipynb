```python
import cv2
import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Load image and reshape to feature space
image = cv2.imread('/content/View-of-robot-workspace.png')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
pixels = image_rgb.reshape((-1, 3))

# Fit GMM
gmm = GaussianMixture(n_components=4, covariance_type='tied').fit(pixels)
labels = gmm.predict(pixels)

# Reshape labels to image
segmented_image = labels.reshape(image.shape[:2])

# Display segmentation
plt.imshow(segmented_image, cmap='viridis')
plt.title('Segmented Robot workspace view')
plt.axis('off')
plt.show()
```

```python
# Display Original image
plt.imshow(image, cmap='viridis')
plt.title('Original Robot workspace view')
plt.axis('off')
plt.show()
```
